```python
import pandas as pd

from neighborhood_functions import search_save_neighborhoods, load_neighborhoods
from global_variables import NEIGHBORHOOD_SIZE, IMAGE_WIDTH, IMAGE_HEIGHT

from files_path_variables import all_cleaned_manually_text_files, all_cleaned_corr_text_files, metadata_file

#Load all metadata needed for filtering
print(f"Dataframe for file {metadata_file}. Loading data...")
metadata_frame = pd.read_csv(metadata_file, sep=',')
metadata_frame = metadata_frame.dropna(how='all', axis='columns')
metadata_frame = metadata_frame.dropna(how='all', axis='rows')
if not metadata_frame.empty:
    metadata_frame = metadata_frame.dropna(subset=["codigo"])
    codes_list = list(metadata_frame['codigo'])
    period_list = list(metadata_frame['periodo'])
    territorial_entity_list = list(metadata_frame['entidad territorial'])
    nationality_list = list(metadata_frame['nacionalidad'])
print("Finished loading dataframe data.")

#Control number saved neighborhoods
number_of_saved_neighborhoods = 0
```

```python
#Search, save or load neighborhoods

print("What do you want to do?")
print("\n(1) Look up for neighborhoods based on a term.")
print("(2) Load saved neighborhoods.\n")
chosen_option = input("")

if chosen_option == "1":
    
    #Get and save all the neighborhoods of a specific term
    term = input("\nWhich term do you want to get the neighborhoods from? : ")
    length_of_neighborhoods = input("\nWhich window size do you want the neighborhoods to be? (default 50) : ")
    if length_of_neighborhoods == "":
        length_of_neighborhoods = NEIGHBORHOOD_SIZE
    elif length_of_neighborhoods.isdigit():
        length_of_neighborhoods = int(length_of_neighborhoods)
    else:
        print("Invalid size value.")
        length_of_neighborhoods = -1

    if length_of_neighborhoods != -1:
        print("\nSearching for the neighborhoods...\n")
        neighborhoods_dictionary, indexes_term_dictionary, start_indexes_dictionary, finish_indexes_dictionary = search_save_neighborhoods(term, length_of_neighborhoods, databases=[all_cleaned_manually_text_files, all_cleaned_corr_text_files]) 

elif chosen_option == "2":

    neighborhoods_dictionary, indexes_term_dictionary, start_indexes_dictionary, finish_indexes_dictionary, length_of_neighborhoods, term = load_neighborhoods()
        
    print(f"\nFile loaded succesfully. Loaded neighborhoods of the term {term} for size {length_of_neighborhoods} from {len(neighborhoods_dictionary)} files.")

else:

    print("\nInvalid option.")
```

```python
#Print some basic information about the neighborhoods

option = input("Print some basic information about the neighborhoods? (yes or no): ")

#make chosen option lowercase
option = option.lower()

if option == "yes" or option == "y":

    print(f"\nNeighborhoods of the term {term} of size {length_of_neighborhoods}:\n")

    amount_of_neighborhoods = 0
    for key in neighborhoods_dictionary.keys():
        temp_amount_neighborhoods = len(neighborhoods_dictionary[key])
        amount_of_neighborhoods += temp_amount_neighborhoods
        print(f"Found {temp_amount_neighborhoods} neighborhoods in document {key}.\n")
        
    print(f"Finished printing the information for {amount_of_neighborhoods} neighborhoods.")

else:
    print("\nNo neighborhood information printed.")
```

```python
#Click functions
from display_pdf_pages_functions import get_pdf_page

def next_button_clicked(next_click):
    global current_document, current_neighborhood, list_of_neighborhoods, amount_of_neighborhoods, pdf_path, start_indexes, neighborhoods_vbox

    print("CN",current_neighborhood, amount_of_neighborhoods)

    if current_neighborhood >= amount_of_neighborhoods-1:
        if current_document < amount_of_documents-1:
            current_document += 1
            current_neighborhood = 0
            list_of_neighborhoods = list(neighborhoods_dictionary[list_of_documents[current_document]])
            amount_of_neighborhoods = len(list_of_neighborhoods)
    else:
        current_neighborhood += 1

    if button_back.disabled == True:
        button_back.disabled = False

    if not button_next.disabled:
        neighborhood_text.value = f"Neighborhood #{current_neighborhood} from {amount_of_neighborhoods} from document {list_of_documents[current_document]}."
        textarea.value = list_of_neighborhoods[current_neighborhood]
        pdf_path = Path(pdf_dir,list_of_documents[current_document]+".pdf")
        start_indexes = start_indexes_dictionary[list_of_documents[current_document]]

    #If it's the last document you can't go back anymore
    if current_document >= amount_of_documents-1 and current_neighborhood >= amount_of_neighborhoods-1:
        button_next.disabled = True
        
    neighborhoods_vbox.children = (metadata_message, textarea_metadata_hbox, next_back_hbox, button_save, button_pdf)

def back_button_clicked(back_click):
    global current_document, current_neighborhood, list_of_neighborhoods, amount_of_neighborhoods, pdf_path, start_indexes, neighborhoods_vbox

    if current_neighborhood == 0:
        if current_document != 0:
            current_document -= 1
            list_of_neighborhoods = list(neighborhoods_dictionary[list_of_documents[current_document]])
            amount_of_neighborhoods = len(list_of_neighborhoods)
            current_neighborhood = amount_of_neighborhoods - 1
    else:
        current_neighborhood -= 1

    if button_next.disabled == True:
        button_next.disabled = False

    if not button_back.disabled:
        neighborhood_text.value = f"Neighborhood #{current_neighborhood} from {amount_of_neighborhoods} from document {list_of_documents[current_document]}."
        textarea.value = list_of_neighborhoods[current_neighborhood]
        pdf_path = Path(pdf_dir,list_of_documents[current_document]+".pdf")
        start_indexes = start_indexes_dictionary[list_of_documents[current_document]]

    #If it's the last document you can't go back anymore
    if current_document == 0:
        button_back.disabled = True
        
    neighborhoods_vbox.children = (metadata_message, textarea_metadata_hbox, next_back_hbox, button_save, button_pdf)

# Define the function to handle confirm button click
def confirm_button_clicked(confirm_click):

    global number_of_saved_neighborhoods, neighborhoods_dictionary, neighborhoods_vbox
                    
    neighborhoods_dictionary[list_of_documents[current_document]][current_neighborhood] = textarea.value

    #Enable again
    disabled = False
    next_back_hbox.disabled = False
    textarea.disabled = False

    save_confirm_message = Text(value="Neighborhood succesfully saved to the variable. If you want to save them to the file, run the next cell.", layout=Layout(width='auto'))

    number_of_saved_neighborhoods += 1

    neighborhoods_vbox.children = (save_confirm_message, metadata_message, textarea_metadata_hbox, next_back_hbox, button_save, button_pdf)

# Define the function to handle cancel button click
def cancel_button_clicked(cancel_click):

    global neighborhoods_vbox
            
    #Enable textarea for further editing
    textarea.disabled = False
    next_back_hbox.disabled = False
    
    neighborhoods_vbox.children = (metadata_message, textarea_metadata_hbox, next_back_hbox, button_save, button_pdf)

# Define a function to handle button click event
def save_modified_neighborhood_to_variable(save_click):

    global neighborhoods_vbox

    #Disable textarea
    textarea.disabled = True
    next_back_hbox.disabled = True

    button_confirm.on_click(confirm_button_clicked)
    button_cancel.on_click(cancel_button_clicked)
        
    neighborhoods_vbox.children = (metadata_message, textarea_metadata_hbox, next_back_hbox, confirm_cancel_hbox)

#Function to search for a pdf page and display the image
def search_pdf_pages(pdf_click):

    global main_image, previous_image, next_image, page_number, text_main_image, previous_page_bool, next_page_bool, neighborhoods_vbox

    print(pdf_click)

    #Check if it's a valid neighborhood
    if current_neighborhood <= len(start_indexes)-1:

        print("CN",current_neighborhood)

        #Search for the pdf page image with the neighborhood and the page image before and after in case the neighborhood overlaps
        pdf_page, page_number, pdf_page_previous, pdf_page_next = get_pdf_page(pdf_path, start_indexes[current_neighborhood], width=IMAGE_WIDTH, height=IMAGE_HEIGHT)

        #If the page was found
        if pdf_page == None:
            print(f"The page with the neighborhood was not found.")
        else:
            main_image.value = open(pdf_page, 'rb').read()
            text_main_image.value = f"Page number {page_number}:"
            previous_page_bool = False
            next_page_bool = False
            if pdf_page_previous != None:
                previous_image.value = open(pdf_page_previous, 'rb').read()
            if pdf_page_next != None:
                next_image.value = open(pdf_page_next, 'rb').read()

        neighborhoods_vbox.children = (metadata_message, textarea_metadata_hbox, next_back_hbox, button_save, button_extra, text_main_image, main_image)

    else:
        print(f"\nThe neighborhood number {number_of_neighborhood} in document {pdf_path.stem} doesn't exist.")

def display_another_image(extra_click):

    global text_previous_image, text_next_image, previous_page_bool, next_page_bool, neighborhoods_vbox

    #If asking for the previous page and that page exists
    if not previous_page_bool and previous_image.value != b"":
        previous_page_bool = True
        text_previous_image.value = f"Page number {page_number-1}:"
        metadata_message, textarea_metadata_hbox, next_back_hbox, button_save, button_pdf
        neighborhoods_vbox.children = (metadata_message, textarea_metadata_hbox, next_back_hbox, button_save, button_extra, text_main_image, main_image, text_previous_image, previous_image)
    #If asking for the next page and that page exists
    elif not next_page_bool and next_image.value != b"":
        next_page_bool = True
        text_next_image.value = f"Page number {page_number+1}:"
        neighborhoods_vbox.children = (metadata_message, textarea_metadata_hbox, next_back_hbox, button_save, button_extra, text_main_image, main_image, text_previous_image, previous_image, text_next_image, next_image)

def get_filtered_dictionaries(df, column, partial_string):
    list_of_documents_filtered = []

    #Sometimes could be None 
    if not isinstance(partial_string, str):
        print(type(partial_string), partial_string)

    print("PS", partial_string)

    filtered_df = df[df[column].str.contains(partial_string, case=False, na=False)]
    list_of_documents_filtered = filtered_df['codigo'].tolist()

    #Get rid of documents that have no neighborhoods
    list_of_documents_filtered = list(set(list_of_documents_filtered).intersection(list_of_documents_unfiltered))
    
    return list_of_documents_filtered

def get_all_filtered_variables(list_of_documents):
    current_document = 0
    current_neighborhood = 0
    amount_of_documents = len(list_of_documents)
    print("AOD FV",amount_of_documents)

    pdf_path = Path(pdf_dir,list_of_documents[current_document]+".pdf")
    start_indexes = start_indexes_dictionary[list_of_documents[current_document]]
    list_of_neighborhoods = list(neighborhoods_dictionary[list_of_documents[current_document]])
    amount_of_neighborhoods = len(list_of_neighborhoods)

    return current_document, current_neighborhood, amount_of_documents, pdf_path, start_indexes, list_of_neighborhoods, amount_of_neighborhoods
    
def change_text_on_widgets(metadata_message, neighborhood_text, textarea):
    metadata_message.value = f"Metadata filters. Periodo: {str(period_filtering_options)}. Entidad Territorial: {str(entity_filtering_options)}. Nacionalidad: {str(nationality_filtering_options)}"

    neighborhood_text.value = f"Neighborhood #{current_neighborhood} from {amount_of_neighborhoods} from document {list_of_documents[current_document]}."

    textarea.value = list_of_neighborhoods[current_neighborhood]

    return metadata_message, neighborhood_text, textarea

def publish_updated_widgets(list_of_documents):

    global current_document, current_neighborhood, amount_of_documents, pdf_path, start_indexes, \
    list_of_neighborhoods, amount_of_neighborhoods, metadata_message, neighborhood_text, textarea, neighborhoods_vbox

    #When deselect the list_of_documents might be empty and give an error so ignore cause it gets updated afterwards
    if list_of_documents:
        current_document, current_neighborhood, amount_of_documents, \
        pdf_path, start_indexes, list_of_neighborhoods, amount_of_neighborhoods = get_all_filtered_variables(list_of_documents)
    
        print("AOD",amount_of_documents)
    
        metadata_message, neighborhood_text, textarea = change_text_on_widgets(metadata_message, neighborhood_text, textarea)

        print(type(metadata_message),type(textarea_metadata_hbox),type(next_back_hbox),type(button_save), type(button_pdf))
    
        neighborhoods_vbox.children = (metadata_message, textarea_metadata_hbox, next_back_hbox, button_save, button_pdf)

def on_nationality_checkbox_change(change):
    global current_document, current_neighborhood, amount_of_documents, pdf_path, start_indexes, \
    list_of_neighborhoods, amount_of_neighborhoods, list_of_documents, nationality_filtering_options

    if change['new']:
        list_of_documents = []
        print(nationality_filtering_options,"F")
        if nationality_filtering_options:
            nationality_filtering_options += ('varios',)
        else: 
            nationality_filtering_options = ('varios',)
        print(nationality_filtering_options,"S")
        for option in nationality_filtering_options: list_of_documents += get_filtered_dictionaries(metadata_frame, 'nacionalidad', option)
        if entity_filtering_options: 
            for option in entity_filtering_options: list_of_documents += get_filtered_dictionaries(metadata_frame, 'entidad territorial', option)
        if period_filtering_options: 
            for option in period_filtering_options: list_of_documents += get_filtered_dictionaries(metadata_frame, 'periodo', option)
        #Needs to filter out the elements that have no neighborhoods and make them unique
        list_of_documents = list(set(list_of_documents).intersection(list_of_documents_unfiltered))

    publish_updated_widgets(list_of_documents)

def on_period_select_multiple_change(change):
    global list_of_documents, list_of_documents_periods, period_filtering_options
    
    period_filtering_options = change['new']
    if period_filtering_options:
        list_of_documents = []
        list_of_documents_periods = []
        for option in period_filtering_options: list_of_documents_periods += get_filtered_dictionaries(metadata_frame, 'periodo', option)
        #Needs to filter out the elements that have no neighborhoods and make them unique
        list_of_documents = list_of_documents_periods + list_of_documents_entities + list_of_documents_nationalities
        list_of_documents = list(set(list_of_documents).intersection(list_of_documents_unfiltered))
        
    publish_updated_widgets(list_of_documents)

def on_entity_accordion_change(change):
    global list_of_documents, list_of_documents_entities, entity_filtering_options
    
    #entity_filtering_options = change['new']
    selected_regions = []
    print(entity_filtering_options, 'FF')

    list_of_documents = []
    list_of_documents_entities = []

    for entity_widget in entity_widgets:
        countries_entity_widget = entity_widget.children[1]
        selected_countries = list(countries_entity_widget.value)
        if selected_countries:
            #selected_regions.append(entity_widget.children[0].value)
            selected_regions.extend(selected_countries)

    entity_filtering_options = selected_regions

    print(entity_filtering_options, 'FF')
    
    for option in entity_filtering_options: list_of_documents_entities += get_filtered_dictionaries(metadata_frame, 'entidad territorial', option)
    
    list_of_documents = list_of_documents_periods + list_of_documents_entities + list_of_documents_nationalities
    #Needs to filter out the elements that have no neighborhoods and make them unique
    list_of_documents = list(set(list_of_documents).intersection(list_of_documents_unfiltered))

    publish_updated_widgets(list_of_documents)

def on_nationality_accordion_change(change):
    global list_of_documents, list_of_documents_nationalities, nationality_filtering_options
    
    #entity_filtering_options = change['new']
    selected_regions = []
    print(nationality_filtering_options, 'FN')

    list_of_documents = []
    list_of_documents_nationalities = []

    for nationality_widget in nationality_widgets:
        countries_nationality_widget = nationality_widget.children[1]
        selected_countries = list(countries_nationality_widget.value)
        if selected_countries:
            #selected_regions.append(nationality_widget.children[0].value)
            selected_regions.extend(selected_countries)

    nationality_filtering_options = selected_regions

    for option in nationality_filtering_options: list_of_documents_nationalities += get_filtered_dictionaries(metadata_frame, 'nacionalidad', option)

    list_of_documents = list_of_documents_periods + list_of_documents_entities + list_of_documents_nationalities
    #Needs to filter out the elements that have no neighborhoods and make them unique
    list_of_documents = list(set(list_of_documents).intersection(list_of_documents_unfiltered))

    publish_updated_widgets(list_of_documents)

def deselect_all_periods(button):

    global list_of_documents_periods, list_of_documents, period_filtering_options, period_select

    period_filtering_options = []
    
    period_select.value = []
    
    list_of_documents_periods = []
    list_of_documents = list_of_documents_entities + list_of_documents_nationalities
    print("Deselect period", len(list_of_documents))

    #If everything is deselected then just use the original list of documents
    if not (list(period_filtering_options) + entity_filtering_options + nationality_filtering_options):
        list_of_documents = copy.deepcopy(list_of_documents_unfiltered)

    publish_updated_widgets(list_of_documents)

def deselect_all_entities(button):

    print("entrz")

    global list_of_documents_entities, list_of_documents, entity_filtering_options, entity_widgets

    entity_filtering_options = []

    for entity_widget in entity_widgets:
        countries_widget = entity_widget.children[1]
        print("E",countries_widget.value)
        countries_widget.value = []  # Deselect all options
        print("E2",countries_widget.value)

    list_of_documents_entities = []
    list_of_documents = list_of_documents_periods + list_of_documents_nationalities

    #If everything is deselected then just use the original list of documents
    if not (list(period_filtering_options) + entity_filtering_options + nationality_filtering_options):
        list_of_documents = copy.deepcopy(list_of_documents_unfiltered)

    publish_updated_widgets(list_of_documents)

def deselect_all_nationalities(button):

    global list_of_documents_nationalities, list_of_documents, nationality_filtering_options, nationality_filtering_options, nationality_checkbox, nationality_widgets

    nationality_filtering_options = []

    nationality_checkbox.value = False
    
    for nationality_widget in nationality_widgets:
        countries_widget = nationality_widget.children[1]
        countries_widget.value = []  # Deselect all options

    list_of_documents_nationalities = []
    list_of_documents = list_of_documents_periods + list_of_documents_entities

    #If everything is deselected then just use the original list of documents
    if not (list(period_filtering_options) + entity_filtering_options + nationality_filtering_options):
        list_of_documents = copy.deepcopy(list_of_documents_unfiltered)

    publish_updated_widgets(list_of_documents)
```

```python
#Loop between the neighborhoods
from ipywidgets import Textarea, Button, VBox, HBox, Text, Layout, Image, interact, SelectMultiple, Accordion, Label, Checkbox
from files_path_variables import pdf_dir
from pathlib import Path
from useful_functions import split_metadata_period, split_metadata_entity, split_metadata_nationality

import copy

list_of_documents_unfiltered = list(neighborhoods_dictionary.keys())
list_of_documents = copy.deepcopy(list_of_documents_unfiltered)
list_of_documents_periods = []
list_of_documents_entities = []
list_of_documents_nationalities = []

#Store filtering option
nationality_filtering_options = []
entity_filtering_options = []
period_filtering_options = []

#Generate all metadata choices
metadata_message = Text(value="Metadata filters.", layout=Layout(width='auto'))

period_label = Text(value="Period filters.", layout=Layout(width='auto'))
entity_label = Text(value="Territorial Entity filters.", layout=Layout(width='auto'))
nationality_label = Text(value="Nationality filters.", layout=Layout(width='auto'))

period_choices = split_metadata_period(period_list)
entity_choices = split_metadata_entity(territorial_entity_list)
nationality_choices = split_metadata_nationality(nationality_list)

button_deselect_periods = Button(description="Deselect All")
button_deselect_entities = Button(description="Deselect All")
button_deselect_nationalities = Button(description="Deselect All")

period_select = SelectMultiple(options=period_choices)
period_vbox = VBox([period_label, period_select, button_deselect_periods])

entity_accordion = Accordion()
entity_widgets = []

for region, countries in entity_choices.items():
    entity_widget = VBox([Label(region), SelectMultiple(options=countries)])
    entity_widgets.append(entity_widget)

entity_accordion.children = entity_widgets
for i, region in enumerate(entity_choices.keys()):
    entity_accordion.set_title(i, region)

entity_vbox = VBox([entity_label, entity_accordion, button_deselect_entities])

nationality_accordion = Accordion()
nationality_widgets = []

for region, countries in nationality_choices.items():
    if countries:
        nationality_widget = VBox([Label(region), SelectMultiple(options=countries)])
        nationality_widgets.append(nationality_widget)
#Add extra checkbox for varios
nationality_checkbox = Checkbox(description='Varios')

nationality_accordion.children = nationality_widgets
for i, region in enumerate(nationality_choices.keys()):
    nationality_accordion.set_title(i, region)
nationality_vbox = VBox([nationality_label, nationality_accordion, nationality_checkbox, button_deselect_nationalities])

metadata_hbox = HBox([period_vbox, entity_vbox, nationality_vbox])

textarea = Textarea(value="", layout={'width': '900px', 'height': '500px'})
button_save = Button(description="Save")
button_next = Button(description="Next")
button_back = Button(description="Back")
button_pdf = Button(description="Show PDF")

#Control display
previous_page_bool = False
next_page_bool = False

#Open first image which is the chosen one
text_main_image = Text(value="", layout=Layout(width='auto'))
main_image = Image(value=b"", format='jpg', width=IMAGE_WIDTH, height=IMAGE_HEIGHT)
text_previous_image = Text(value="", layout=Layout(width='auto'))
previous_image = Image(value=b"", format='jpg', width=IMAGE_WIDTH, height=IMAGE_HEIGHT)
text_next_image = Text(value="", layout=Layout(width='auto'))
next_image = Image(value=b"", format='jpg', width=IMAGE_WIDTH, height=IMAGE_HEIGHT)
       
# Create a button widget for extra pdf pages
button_extra = Button(description='Extra Page')

#Subbuttons after Saving
button_confirm = Button(description="Confirm Save")
button_cancel = Button(description="Cancel Save")

amount_of_documents = len(list_of_documents)
current_document = 0
pdf_path = Path(pdf_dir,list_of_documents[current_document]+".pdf")
start_indexes = start_indexes_dictionary[list_of_documents[current_document]]
list_of_neighborhoods = list(neighborhoods_dictionary[list_of_documents[current_document]])
amount_of_neighborhoods = len(list_of_neighborhoods)
current_neighborhood = 0

neighborhood_text = Text(value=f"Neighborhood #{current_neighborhood} from {amount_of_neighborhoods} from document {list_of_documents[current_document]}.", layout=Layout(width='auto'))
textarea.value = list_of_neighborhoods[current_neighborhood]

textarea_vbox = VBox([neighborhood_text, textarea])
textarea_metadata_hbox = HBox([textarea_vbox, metadata_hbox])

next_back_hbox = HBox([button_back, button_next])
confirm_cancel_hbox = HBox([button_confirm, button_cancel])
neighborhoods_vbox = VBox([metadata_message, textarea_metadata_hbox, next_back_hbox, button_save, button_pdf])

display(neighborhoods_vbox)

button_next.on_click(next_button_clicked)
button_back.on_click(back_button_clicked)
button_save.on_click(save_modified_neighborhood_to_variable)
button_pdf.on_click(search_pdf_pages)
button_extra.on_click(display_another_image)

#Metadata widgets calls
nationality_checkbox.observe(on_nationality_checkbox_change, names='value')
period_select.observe(on_period_select_multiple_change, names='value')

for entity_widget in entity_widgets:
    countries_entity_widget = entity_widget.children[1]
    countries_entity_widget.observe(on_entity_accordion_change, 'value')
for nationality_widget in nationality_widgets:
    countries_nationality_widget = nationality_widget.children[1]
    countries_nationality_widget.observe(on_nationality_accordion_change, 'value')

button_deselect_periods.on_click(deselect_all_periods)
button_deselect_entities.on_click(deselect_all_entities)
button_deselect_nationalities.on_click(deselect_all_nationalities)
```

```python
#Add term to the extra corrections file
import csv
from files_path_variables import path_to_rules_file

term_to_correct = input("What term you want to correct? ")
new_term = input("\nWhat new term you want to enter? ")

confirmation = input(f"\nAre you sure you want to correct the term {term_to_correct} to {new_term}? (default yes)")

print(confirmation)

if confirmation == "yes" or confirmation == "y" or confirmation == "":

    # Open the CSV file in read mode
    with open(path_to_rules_file, 'r') as file:
        # Read the existing data from the CSV file
        reader = csv.reader(file)
        rows = list(reader)  # Convert the reader object to a list

    # Add the new value and key to the rules data
    new_key = term_to_correct
    new_value = new_term
    new_row = [new_key, new_value]
    rows.append(new_row)

    # Open the same CSV file in write mode
    with open(path_to_rules_file, 'w', newline='') as file:
        # Write the updated rules data back to the CSV file
        writer = csv.writer(file)
        writer.writerows(rows)
        
    print(f"\nNew correction *{term_to_correct}:{new_term}* correctly saved.")
        
else:
    print("New correction not saved.")
```
